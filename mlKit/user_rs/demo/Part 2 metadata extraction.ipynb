## For track information 

```python
import pandas as pd

df = pd.read_csv("../data/tmp/seeds.csv")
```

```python
# place SQL results in a DataFrame - method 2: pandas read_sql_query()
import sqlite3 as sql
import pandas as pd
connection = sql.Connection("../data/input/track_metadata.db")
query = 'SELECT * FROM songs'
meta = pd.read_sql_query(query, con=connection)
```

```python
meta = meta[meta.track_id.isin(df.track_id)].reset_index(drop=True)

# drop last three columns
meta = meta.iloc[:, :-3]

meta.head()
```

```python
# save metadata
meta.to_csv("cleaned_data/tmp/tracks_metadata.csv", index=False)
```
